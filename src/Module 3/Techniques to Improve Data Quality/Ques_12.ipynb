```python
# Activity 2: Dealing with Duplicates & Redundancy

# Task A: Identifying Duplicate Records

import pandas as pd

def load_dataset(file_path):
    """ Load dataset and handle file loading errors. """
    try:
        df = pd.read_csv(file_path)
        return df
    except FileNotFoundError:
        print(f"Error: The file at {file_path} was not found.")
        return None
    except pd.errors.EmptyDataError:
        print("Error: The file is empty.")
        return None
    except Exception as e:
        print(f"An unexpected error occurred while loading the file: {e}")
        return None

# Task 7: Identify Complete Duplicates
def identify_complete_duplicates(df):
    """ Identify and return complete duplicate rows in the DataFrame. """
    if df is not None:
        try:
            duplicates = df[df.duplicated()]
            if duplicates.empty:
                print("No complete duplicates found.")
            else:
                print(f"Found {len(duplicates)} complete duplicate rows.")
            return duplicates
        except Exception as e:
            print(f"An error occurred while identifying complete duplicates: {e}")
            return None
    else:
        print("DataFrame is empty or failed to load.")
        return None

# Task 8: Identify Duplicates Based on Specific Columns
def identify_duplicates_by_columns(df, columns):
    """ Identify duplicates based on specific columns. """
    if df is not None:
        try:
            if not set(columns).issubset(df.columns):
                print("Error: One or more specified columns do not exist in the DataFrame.")
                return None
            duplicates = df[df.duplicated(subset=columns)]
            if duplicates.empty:
                print("No duplicates found based on specified columns.")
            else:
                print(f"Found {len(duplicates)} duplicate rows based on columns {columns}.")
            return duplicates
        except Exception as e:
            print(f"An error occurred while identifying duplicates based on columns: {e}")
            return None
    else:
        print("DataFrame is empty or failed to load.")
        return None

# Task 9: Count Duplicate Rows
def count_duplicate_rows(df):
    """ Count and return the number of duplicate rows in the DataFrame. """
    if df is not None:
        try:
            duplicate_count = df.duplicated().sum()
            print(f"Total number of duplicate rows: {duplicate_count}")
            return duplicate_count
        except Exception as e:
            print(f"An error occurred while counting duplicate rows: {e}")
            return None
    else:
        print("DataFrame is empty or failed to load.")
        return None

# Example usage:
file_path = 'employees.csv'  # Replace with actual file path
df = load_dataset(file_path)

# Identifying complete duplicates
identify_complete_duplicates(df)

# Identifying duplicates based on specific columns (e.g., 'email', 'employee_id')
identify_duplicates_by_columns(df, ['email', 'employee_id'])

# Counting total number of duplicate rows
count_duplicate_rows(df)
```

```python
# Task B: Deduplication Techniques

import pandas as pd

def load_dataset(file_path):
    """ Load dataset and handle file loading errors. """
    try:
        df = pd.read_csv(file_path)
        return df
    except FileNotFoundError:
        print(f"Error: The file at {file_path} was not found.")
        return None
    except pd.errors.EmptyDataError:
        print("Error: The file is empty.")
        return None
    except Exception as e:
        print(f"An unexpected error occurred while loading the file: {e}")
        return None

# Task 10: Remove Complete Duplicates
def remove_complete_duplicates(df):
    """ Remove complete duplicate rows and keep the first occurrence. """
    if df is not None:
        try:
            df_cleaned = df.drop_duplicates(keep='first')
            print(f"Removed complete duplicates. Remaining rows: {len(df_cleaned)}")
            return df_cleaned
        except Exception as e:
            print(f"An error occurred while removing complete duplicates: {e}")
            return None
    else:
        print("DataFrame is empty or failed to load.")
        return None

# Task 11: Subset Deduplication
def remove_duplicates_by_subset(df, subset_columns):
    """ Remove duplicates based on a subset of columns. """
    if df is not None:
        try:
            if not set(subset_columns).issubset(df.columns):
                print("Error: One or more specified columns do not exist in the DataFrame.")
                return None
            df_cleaned = df.drop_duplicates(subset=subset_columns, keep='first')
            print(f"Removed duplicates based on columns {subset_columns}. Remaining rows: {len(df_cleaned)}")
            return df_cleaned
        except Exception as e:
            print(f"An error occurred while removing duplicates by subset: {e}")
            return None
    else:
        print("DataFrame is empty or failed to load.")
        return None

# Task 12: Keep Last Occurrence
def remove_duplicates_keep_last(df):
    """ Remove duplicates but keep the last occurrence in the dataset. """
    if df is not None:
        try:
            df_cleaned = df.drop_duplicates(keep='last')
            print(f"Removed duplicates and kept the last occurrence. Remaining rows: {len(df_cleaned)}")
            return df_cleaned
        except Exception as e:
            print(f"An error occurred while keeping the last occurrence of duplicates: {e}")
            return None
    else:
        print("DataFrame is empty or failed to load.")
        return None

# Example usage:
file_path = 'employees.csv'  # Replace with actual file path
df = load_dataset(file_path)

# Remove complete duplicates
df_cleaned_complete = remove_complete_duplicates(df)

# Remove duplicates based on specific columns (e.g., 'email', 'employee_id')
df_cleaned_subset = remove_duplicates_by_subset(df, ['email', 'employee_id'])

# Keep the last occurrence of duplicates
df_cleaned_last = remove_duplicates_keep_last(df)
```
