```python
# Data Profiling Techniques Examples

# 1. Descriptive Statistics:
# Task 1: Calculate the mean, median, and mode for sales figures in a retail dataset.
# Task 2: Analyze the average age, median, and mode in a customer demographic
# dataset.
# Task 3: Determine the mean, median, and mode of daily website visit counts.


import pandas as pd
from statistics import mode

# Sample data
data = {
    'sales': [200, 220, 250, 200, 300],
    'age': [25, 30, 30, 35, 40],
    'visits': [100, 120, 120, 140, 160]
}
df = pd.DataFrame(data)

# Task 1: Sales statistics
print("Sales - Mean:", df['sales'].mean())
print("Sales - Median:", df['sales'].median())
print("Sales - Mode:", mode(df['sales']))

# Task 2: Age statistics
print("\nAge - Mean:", df['age'].mean())
print("Age - Median:", df['age'].median())
print("Age - Mode:", mode(df['age']))

# Task 3: Daily visits statistics
print("\nVisits - Mean:", df['visits'].mean())
print("Visits - Median:", df['visits'].median())
print("Visits - Mode:", mode(df['visits']))




# 2. Distribution Analysis:
# Task 1: Visualize the distribution of product prices using histograms.
# Task 2: Create a distribution chart for exam scores in an educational dataset.
# Task 3: Analyze the distribution of order quantities for inventory management.


import matplotlib.pyplot as plt

# Task 1: Distribution of product prices
plt.hist([10, 20, 20, 30, 40, 50, 60, 70], bins=6, color='skyblue', edgecolor='black')
plt.title('Product Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Task 2: Exam score distribution
exam_scores = [55, 60, 65, 70, 70, 75, 80, 85, 90]
plt.hist(exam_scores, bins=5, color='lightgreen', edgecolor='black')
plt.title('Exam Score Distribution')
plt.xlabel('Score')
plt.ylabel('Count')
plt.show()

# Task 3: Order quantity distribution
order_qty = [1, 2, 2, 2, 3, 3, 4, 5, 10]
plt.hist(order_qty, bins=5, color='salmon', edgecolor='black')
plt.title('Order Quantity Distribution')
plt.xlabel('Quantity')
plt.ylabel('Count')
plt.show()




# 3. Data Type Validation:
# Task 1: Validate numeric fields in a dataset to ensure they contain only numbers.
# Task 2: Check for valid date formats in a transaction log.
# Task 3: Validate email formats in a customer contact dataset.

import re
from pandas.api.types import is_numeric_dtype

# Sample data
data_valid = {
    'amount': [100, 150, 200, '300', 'bad'],
    'transaction_date': ['2023-01-01', '2023-02-01', 'bad-date', '2023-03-01', ''],
    'email': ['test@example.com', 'bad-email', 'user@domain.org', 'hello@', None]
}
df_valid = pd.DataFrame(data_valid)

# Task 1: Validate numeric fields
df_valid['amount_numeric'] = pd.to_numeric(df_valid['amount'], errors='coerce')
print("\nNon-numeric entries in 'amount':")
print(df_valid[df_valid['amount_numeric'].isna()])

# Task 2: Validate date format
df_valid['valid_date'] = pd.to_datetime(df_valid['transaction_date'], errors='coerce')
print("\nInvalid dates in 'transaction_date':")
print(df_valid[df_valid['valid_date'].isna()])

# Task 3: Validate email format
def is_valid_email(email):
    return bool(re.match(r"[^@]+@[^@]+\.[^@]+", str(email)))

df_valid['valid_email'] = df_valid['email'].apply(is_valid_email)
print("\nInvalid emails:")
print(df_valid[~df_valid['valid_email']])






```
